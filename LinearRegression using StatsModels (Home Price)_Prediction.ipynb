```python
#Removing the $ sign
df['MedianHomePrice']=df['MedianHomePrice'].replace({'\$':''}, regex=True)
df
```

```python
df['MedianHomePrice']=df['MedianHomePrice'].replace({',':''}, regex=True)
df                                                   
```

```python
#pd.to_numeric(df)
df['MedianHomePrice']=pd.to_numeric(df['MedianHomePrice'])
df.info()
```

```python
var1=smf.ols(formula='MedianHomePrice~RateOfInterest', data=df)
var2=var1.fit()
```

```python
print(var2.summary())
```

```python
X_new=pd.DataFrame({'RateOfInterest':[7]})
var2.predict(X_new)
```
